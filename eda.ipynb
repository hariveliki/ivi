```python
import pandas as pd
```

```python
df = pd.read_csv('data/covid_arg.csv')
```

```python
df.info()
```

```python
df.head(5)
```

```python
gender_mapping = {"F": "Female", "M":"Male", "NR": "Unknown"}
# change column sexo to gender for df
df.rename({"sexo": "gender"}, axis=1, inplace=True)
df.rename({"edad": "age"}, axis=1, inplace=True)
df.rename({"residencia_provincia_nombre": "province"}, axis=1, inplace=True)
# change all values from column gender to specified mapping
df["gender"] = df["gender"].map(gender_mapping)
```

```python
df.columns
```

```python
# value counts edad
print(df.age.value_counts())
# max edad
print(df.age.max())
print(df.age.min())
# remove values < 0 and > 120
df = df[df.age > 0]
df = df[df.age < 100]
```

```python
# save first N rows of df to csv
df.sample(frac=0.1).to_csv('data/covid_arg_0_1.csv')
```
